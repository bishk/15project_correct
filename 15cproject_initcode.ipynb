```python
import librosa
from librosa import display
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
from IPython.display import Audio
```

```python
def fft_wrapper(file_name, show_plot = True, n_secs = 90, f_max_plot = 2000, pass_title = None):
    sound, samplerate = librosa.load(file_name, sr = None)
    n_frames = n_secs*samplerate
    
    yf = np.fft.fft(sound[:n_frames])
    fs = np.fft.fftfreq(n_frames, 1/samplerate)
    real_scaling = 2.0/n_frames * np.abs(yf[:n_frames//2]) #taking real components, appropriately scaling
    fs_cut = fs[:n_frames//2]
    #T = 1.0 / samplerate
    #yf = scipy.fftpack.fft(sound, n = 1000000)
    #xf = np.linspace(0.0, 1.0/(2.0*T), n//2)
    print("fft was complete")
    
    if show_plot:
        %matplotlib inline
        fig, ax = plt.subplots()
        ax.plot(fs_cut, real_scaling)
        plt.xlim((0, f_max_plot))
        plt.ylabel("amplitude (AU)")
        plt.xlabel("frequency (inverse seconds)")
        if pass_title is not None:
            plt.title(pass_title)
        plt.show()
    
    return real_scaling, fs_cut, samplerate
```

```python
PPs = ["PP_1.wav", "PP_2.wav", "PP_3.wav", "hamlet.wav"] #recordings of the same passage of pride and prejudice
titles = ["Pride and Prejudice 1", "Pride and Prejudice 2", "Pride and Prejudice 3", "Hamlet"]
n = 2160000 #length for this sample length
ffts = np.zeros((len(PPs), n))

for i in np.arange(len(PPs)):
    fft_i, fs, _ = fft_wrapper(PPs[i], pass_title = titles[i])
    ffts[i, :] = fft_i
    
_ = fft_wrapper('schubert.wav', pass_title = "Schubert quintet", f_max_plot = 2000)
```

```python
speaking_freq = fs < 2000
np.corrcoef(ffts[:, speaking_freq])
```

```python
def comp_2(f1, f2):
    _, axs = plt.subplots()
    edges = np.linspace(-15, -5, 100)
    plt.hist2d(np.log(f1), np.log(f2), edges)
    axs.set_aspect('equal', 'box')
    
comp_2(ffts[0, speaking_freq], ffts[1, speaking_freq])
plt.xlabel('log PP reading 1 spectrum')
plt.ylabel('log PP reading 2 spectrum')
plt.title('Correlations in log power spectra')
plt.show()

comp_2(ffts[0, speaking_freq], ffts[3, speaking_freq])
plt.xlabel('log PP reading 1 spectrum')
plt.ylabel('log hamlet spectrum')
plt.title('Correlations in log power spectra')
plt.show()
```

```python
def spectrogram_wrapper(file_name, window_length = 30):
    sound, samplerate = librosa.load(file_name, sr = None)
    
    f, t, Sxx = scipy.signal.spectrogram(sound, samplerate, nperseg = window_length*samplerate, mode = 'magnitude')
    plt.figure()
    plt.pcolormesh(t, f, np.log(Sxx))
    plt.ylim((0, 4000))
    plt.ylabel('frequency (Hz)')
    plt.xlabel('time (sec)')
    
    return f, t, Sxx
```

```python
f, t, Sxx = spectrogram_wrapper('convo.wav', 60)
plt.title('Spectrogram of conversation')
```

```python
Sxx_low = Sxx[f <2000, :]
```

```python
corrs = np.corrcoef(Sxx_low.T)
corrs[corrs > .99] = np.nan #to make colors easier to see
plt.imshow(corrs)
plt.colorbar()
plt.title('Spectrogram temporal correlations')
plt.xlabel('Sequential time intervals (1 min each)')
plt.ylabel('Sequential time intervals (1 min each)')
```
