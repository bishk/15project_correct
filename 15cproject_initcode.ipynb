```python
import librosa
from librosa import display
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
from IPython.display import Audio
```

```python
def plot(file_name):
    sound, samplerate = librosa.load(file_name, sr = None)
    %matplotlib inline
    plt.figure() 
    #librosa.display.waveplot(y = sound, sr = samplerate) 
    #plt.xlabel("time") 
    #plt.ylabel("amplitude") 
    #plt.show()
    n = len(sound)
    T = 1.0 / samplerate
    yf = scipy.fftpack.fft(sound, n = 1000000)
    print("fft was complete")
    xf = np.linspace(0.0, 1.0/(2.0*T), n//2)
    fig, ax = plt.subplots()
    ax.plot(xf, 2.0/n * np.abs(yf[:n//2]))
    plt.ylabel("amplitude")
    plt.xlabel("frequency (inverse seconds)")
    plt.show()
```

```python
plot("PP_1.wav")
plot("/Users/abhishek/Downloads/file_example_WAV_1MG.wav")
```
